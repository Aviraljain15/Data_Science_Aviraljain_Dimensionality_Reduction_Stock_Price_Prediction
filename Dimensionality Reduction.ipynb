# Part 1: Dimensionality Reduction

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
```

```python
print(X.shape)
```

## Standardizing the Data for PCA

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
print(X_pca.shape)
```

## Reduced Data set

```python
pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
pca_df['species'] = y
print(pca_df.head())
```

## Scatter plot 

```python
plt.figure(figsize=(8,6))

plt.scatter(
    pca_df['PC1'],
    pca_df['PC2'],
    c=pca_df['species'],
    cmap='viridis'
)

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Visualization of Iris Dataset')

plt.show()
```

### Observations

- The PCA plot reduces the original 4-dimensional Iris dataset into 2 principal components for visualization.
- Three clusters are visible, representing the three Iris species.
- One cluster (Setosa) is clearly separated from the other two.
- The other two species (Versicolor and Virginica) show slight overlap, indicating similar feature characteristics.
- PCA successfully preserves the overall structure of the dataset while reducing dimensionality.
